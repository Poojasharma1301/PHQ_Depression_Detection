```python
import itertools

# Define the possible scores
scores = [1, 2, 3]

# Generate all combinations using itertools.product
combinations = list(itertools.product(scores, repeat=9))

# Print the combinations
for combo in combinations:
    print(combo)
```

```python
import itertools

# Define the possible scores
scores = [1, 2, 3]

# Generate all combinations using itertools.product
combinations = list(itertools.product(scores, repeat=9))

# Print the combinations with row scores
for combo in combinations:
    row_score = sum(combo)
    combo_with_score = combo + (row_score,)
    print(combo_with_score)
```

```python
import itertools
import csv
import random

# Define the possible scores
scores = [0,1, 2, 3]

# Generate all combinations using itertools.product
combinations = list(itertools.product(scores, repeat=9))

# Create and open a CSV file for writing
with open('Dataset.csv', 'w', newline='') as csvfile:
    csv_writer = csv.writer(csvfile)

    # Write the header row
    header = ['Age', 'Gender', 'Question 1', 'Question 2', 'Question 3', 'Question 4', 'Question 5', 'Question 6', 'Question 7', 'Question 8', 'Question 9', 'Total Score']
    csv_writer.writerow(header)

    # Write the combinations with random age, gender, and row scores
    for combo in combinations:
        row_score = sum(combo)
        age = random.randint(18, 65)  # Generate a random age between 18 and 65
        gender = random.choice(['Male', 'Female'])  # Randomly select gender

        combo_with_age_gender = [age, gender] + list(combo) + [row_score]
        csv_writer.writerow(combo_with_age_gender)

print("CSV file 'Dataset.csv' has been created.")
```
